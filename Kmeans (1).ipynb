```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Erzeugung von zufälligen Datenpunkten
X = np.array([[1, 2], [5, 8], [1.5, 1.8], [8, 8], [1, 0.6], [9, 11]])
```

```python
# Initialisierung des KMeans-Modells mit der Anzahl der Cluster (z.B. 2)
kmeans = KMeans(n_clusters=2)

# Training des Modells mit den Daten
kmeans.fit(X)

# Vorhersage der Cluster-Zugehörigkeit für die Datenpunkte
predicted_labels = kmeans.predict(X)
```

```python
# Extrahierung der Cluster-Zentren
centers = kmeans.cluster_centers_

# Visualisierung der Cluster und Cluster-Zentren
plt.scatter(X[:, 0], X[:, 1], c=predicted_labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.9)
plt.show()
```
